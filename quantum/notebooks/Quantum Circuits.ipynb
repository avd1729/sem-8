```python
# Import necessary libraries
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere
import matplotlib.pyplot as plt
from qiskit.quantum_info import Statevector
```

```python
# Use %matplotlib notebook to enable interactive plotting
%matplotlib inline
```

**OR GATE**

```python
def or_truth_table():
    print("\nTruth Table for OR Gate:")
    print("A B | C (Result)")
    for a in [0, 1]:
        for b in [0, 1]:
            # OR gate result using | (OR operator)
            c = a | b
            print(f"{a} {b} | {c}")

or_truth_table()
```

```python
# Create a Quantum Circuit with 2 qubits (simple OR gate)
qc = QuantumCircuit(1)

# Apply X gates to simulate OR gate logic (set qubits to |1> state)

qc.h(0)

qc.draw('mpl')
```

```python
statevector = Statevector.from_instruction(qc)
```

```python
plot_state_qsphere(statevector.data)
```

```python
plot_bloch_multivector(statevector.data)
```

```python
print("\nQuantum State (Statevector) After OR Gate Logic:")
print(statevector)
```

```python
# Calculate the probabilities of each state after measurement
probabilities = statevector.probabilities_dict()

# Print the measurement probabilities
print("\nMeasurement probabilities:")
print(probabilities)
```

**AND GATE**

```python
def and_truth_table():
    print("\nTruth Table for AND Gate:")
    print("A B | C (Result)")
    for a in [0, 1]:
        for b in [0, 1]:
            # AND gate result using & (AND operator)
            c = a & b
            print(f"{a} {b} | {c}")

and_truth_table()
```

```python
# Create a Quantum Circuit with 3 qubits
qc = QuantumCircuit(3)

# Initialize inputs: Set qubits A and B to |1>
qc.x(0)  # Qubit A = 1

qc.x(1)  # Qubit B = 1

# Apply the Toffoli (CCX) gate: Controlled-Controlled-NOT
qc.ccx(0, 1, 2)  # If both qubits are 1, flip qubit C
# Draw the quantum circuit
qc.draw('mpl')
```

```python
# Simulate the quantum state using Statevector
statevector = Statevector.from_instruction(qc)
```

```python
# Plot the state on the QSphere and Bloch Multivector
plot_state_qsphere(statevector)
```

```python
plot_bloch_multivector(statevector)
```

```python
# Print the quantum state
print("\nQuantum State (Statevector) After AND Gate Logic:")
print(statevector)
```

```python
# Calculate and print measurement probabilities
probabilities = statevector.probabilities_dict()
print("\nMeasurement Probabilities:")
print(probabilities)
```

**NAND GATE**

```python
def nand_truth_table():
    print("\nTruth Table for NAND Gate:")
    print("A B | C (Result)")
    for a in [0, 1]:
        for b in [0, 1]:
            # NAND gate result using the negation of AND
            c = ~(a & b) & 1  # Mask to keep result within 0 and 1
            print(f"{a} {b} | {c}")

nand_truth_table()
```

```python
# Create a Quantum Circuit with 3 qubits
qc = QuantumCircuit(3)
```

```python
# Initialize inputs: Set qubits A and B to |1>
qc.x(0)  # Qubit A = 1
qc.x(1)  # Qubit B = 1

# Apply the Toffoli (CCX) gate: Controlled-Controlled-NOT
qc.ccx(0, 1, 2)  # If both qubits are 1, flip qubit C

# Apply an X gate to the output qubit to get NAND behavior
qc.x(2)  # NOT operation on the AND result to get NAND

# Draw the quantum circuit
qc.draw('mpl')
```

```python
# Simulate the quantum state using Statevector
statevector = Statevector.from_instruction(qc)
```

```python
# Plot the state on the QSphere and Bloch Multivector
plot_state_qsphere(statevector)
```

```python
plot_bloch_multivector(statevector)
```

```python
# Print the quantum state
print("\nQuantum State (Statevector) After NAND Gate Logic:")
print(statevector)
```

```python
# Calculate and print measurement probabilities
probabilities = statevector.probabilities_dict()
print("\nMeasurement Probabilities:")
print(probabilities)
```

**NOR GATE**

```python
# Function to print the truth table for NOR
def nor_truth_table():
    print("\nTruth Table for NOR Gate:")
    print("A B | C (Result)")
    for a in [0, 1]:
        for b in [0, 1]:
            # NOR gate result using the negation of OR
            c = ~(a | b) & 1  # Mask to keep result within 0 and 1
            print(f"{a} {b} | {c}")
nor_truth_table()
```

```python
# Create a Quantum Circuit with 3 qubits
qc = QuantumCircuit(3)

# Initialize inputs: Set qubits A and B to |1>
qc.x(0)  # Qubit A = 1
qc.x(1)  # Qubit B = 1
qc.cx(0, 2)  
qc.cx(1, 2)
qc.x(2)

qc.draw('mpl')
```

```python
# Simulate the quantum state using Statevector
statevector = Statevector.from_instruction(qc)
```

```python
# Plot the state on the QSphere and Bloch Multivector
plot_state_qsphere(statevector)
```

```python
plot_bloch_multivector(statevector)
```

```python

# Print the quantum state
print("\nQuantum State (Statevector) After XOR Gate Logic:")
print(statevector)
```
